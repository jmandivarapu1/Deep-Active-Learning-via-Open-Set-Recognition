```python
def read_the_subfolder_acc(env_name):
    dirs=['40']#'10','15','20','25','30','35',
    accuracy=[]
    for each_dir in dirs:
        crr='/mnt/iscsi/data/Jay/ActiveLearning/'+str(env_name)+'/'+each_dir+'/'
        with open (crr+'acc_'+str(7), 'rb') as fp:accuracy=pickle.load(fp)
 
    return accuracy




```
